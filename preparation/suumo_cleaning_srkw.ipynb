```python
import pandas as pd 
import numpy as np
```

```python
# アクセス　駅・沿線分割(必要？)＋徒歩分数単位切り
access_inf = pd.read_csv('アクセス.csv',dtype='object')
access_inf['駅名'] = access_inf['駅'].str.split('/')
access_inf['駅'] = access_inf['駅名'].str[1]
access_inf['沿線'] = access_inf['駅名'].str[0]

# access_inf['徒歩分数(分)'] = access_inf['徒歩分数'].str[1:-1].astype(int)
access_inf['徒歩分数(分)'] = access_inf['徒歩分数'].str[1:-1].apply(lambda x: int(x) if len(x)<=3 else np.NAN)  #澤柳編集

access_inf['連番'] = access_inf.groupby(['物件CD']).cumcount()+1

access_inf = access_inf.reindex(columns=['物件CD','連番','駅','沿線','徒歩分数(分)'])
access_inf.head()
```

```python
# 周辺環境　距離単位切り
env_inf = pd.read_csv('周辺環境.csv',dtype='object')
env_inf['距離(m)'] = env_inf['距離'].str[:-1].astype(int)
env_inf['連番'] = env_inf.groupby(['物件CD']).cumcount()+1
env_inf = env_inf.reindex(columns=['物件CD','連番','施設名','ジャンル','距離(m)'])
env_inf.head(10)
```

```python
# エクスポート
access_inf.to_csv('アクセス_処理後.csv',index=False,encoding='utf-8_sig')
env_inf.to_csv('周辺環境_処理後.csv',index=False,encoding='utf-8_sig')
```
