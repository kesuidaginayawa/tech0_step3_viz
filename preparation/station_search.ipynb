```python
import requests
import pandas as pd
```

```python
#路線
URL = 'http://express.heartrails.com/api/json?method=getLines'
params = {
    'prefecture':'神奈川県'
}
res = requests.get(URL,params)
line_jsn = res.json()
lines = line_jsn['response']['line']
```

```python
# 駅

URL = 'http://express.heartrails.com/api/json?method=getStations'

stations = [] 

for line in lines:
    params = {
        'line':line
    }
    res = requests.get(URL,params)
    sttn_jsn = res.json()

    stations.append(sttn_jsn)
```

```python
#データフレーム作成

t_x = []
t_y = []
t_line = []
t_name = []
for station in stations:
    for j in station['response']['station']:
        if j['prefecture']=='神奈川県':
            t_line.append(j['line'])
            t_name.append(j['name'])
            t_x.append(float(j['x']))
            t_y.append(float(j['y']))
df = pd.DataFrame({
    '沿線':t_line,
    '駅':t_name,
    'lon':t_x,
    'lat':t_y
})

```

```python
df.to_csv('駅_緯度経度.csv',index=False,encoding='utf-8_sig')
df
```
